```python
#inner join samples_check on site_info by profile to understand how where the highest density of samples were taken
merged_df = pd.merge(samples_check, site_info, on='profile', how='inner')
print(f"Merged dataframe has {len(merged_df)} rows")
merged_df.head()
```

```python
samples_per_location = merged_df['district'].value_counts().sort_values(ascending=False)
print(samples_per_location.head(10))  # Top 10 locations
```

```python
samples_per_district_df = samples_per_district.reset_index()
samples_per_district_df.columns = ['District', 'Number of Samples']

fig = px.bar(
    samples_per_district_df,
    x='District',
    y='Number of Samples',
    title='Number of Soil Samples per District',
    text='Number of Samples'
)

fig.update_traces(textposition='outside')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
merged_df = merged_df.rename(columns={'X_coord': 'lon', 'Y_coord': 'lat'})
```

```python
merged_df[['sample_id', 'lat', 'lon']].dropna().head()
```

```python
fig = px.scatter_mapbox(
    merged_df,
    lat='lat',
    lon='lon',
    hover_name='sample_id',
    color='district',
    zoom=6,
    title='Map of Soil Samples by District',
    height=600
)

fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})
fig.show()
```
